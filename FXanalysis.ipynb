```python
import numpy as np
import pandas as pd
pd.core.common.is_list_like = pd.api.types.is_list_like
import pandas_datareader.data as web
```

```python
##Deutsche Bundesbank Data Repository
#FX_USD = web.DataReader('BIS/US_DAUAUDA', 'quandl')
## Bank of England Official Statistics
FX_USDAUD = web.DataReader('BOE/XUDLADD','quandl')
FX_GBPAUD = web.DataReader('BOE/XUDLADS','quandl')
FX_USDEUR = web.DataReader('BOE/XUDLERD','quandl')
##South African Rand
FX_USDZAR = web.DataReader('BOE/XUDLZRD','quandl')
```

```python
FX_USDAUD.head(),FX_GBPAUD.head(),FX_USDEUR.head(),FX_USDZAR.head()
```

```python
len(FX_USDAUD),len(FX_GBPAUD),len(FX_USDEUR),len(FX_USDZAR)
```

```python
FX_EURAUD = (1.0/FX_USDEUR)*FX_USDAUD
```

```python
FX_EURAUD.head()
```

```python
FX_ZARAUD = (1.0/FX_USDZAR)*FX_USDAUD
```

```python
FX_ZARAUD.head()
```

```python
FX_ZARAUD.plot(figsize=(8, 5),legend='AUD for 1 ZAR')
FX_EURAUD.plot(figsize=(8, 5))
FX_USDAUD.plot(figsize=(8, 5))
FX_GBPAUD.plot(figsize=(8, 5))
FX = pd.concat([FX_ZARAUD,
                FX_EURAUD,
                FX_USDAUD,
                FX_GBPAUD],
               axis=1)
FX.columns = ['AUD/ZAR',
              'AUD/EUR',
              'AUD/USD',
              'AUD/GBP']
FX.plot()
FX.head()
```
